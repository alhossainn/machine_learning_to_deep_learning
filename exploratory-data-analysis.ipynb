## Find the Categorical and Numerical Columns

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
print("Numerical Columns:")
print(list(numerical_cols))

categorical_cols = df.select_dtypes(include=['object']).columns
print("\nCategorical Columns:")
print(list(categorical_cols))
```

## Data Normalization
We scale (normalize) the data to ensure that all numerical features are on the same scale, typically between 0 and 1. This is important because many machine learning algorithms are sensitive to the magnitude of values; features with larger ranges can dominate those with smaller ranges and negatively affect model performance. Normalization helps improve training stability, speeds up convergence, and often leads to better and more reliable model results.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
df.head()
```

## Find Unique Values in Categorical Columns

```python
categorical_cols = df.select_dtypes(include=['object']).columns
print("\nUnique values in categorical columns:")

for col in categorical_cols:
    unique_values = df[col].unique()
    print(f"\n{col}: {len(unique_values)} unique values")
    print(unique_values)
```

## Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

df_label_encoded = df.copy()
le = LabelEncoder()

for col in df_label_encoded.select_dtypes(include=['object']).columns:
    df_label_encoded[col] = le.fit_transform(df_label_encoded[col])

df_label_encoded.head()
```
